```python
# https://auto.gluon.ai/stable/tutorials/tabular/tabular-quick-start.html
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train = TabularDataset('merged_data.csv')
```

```python
train_size = int(len(train) * 0.8)
seed = 1080
# train / test split
train_set = train.sample(train_size,random_state=seed)
test_set = train.drop(train_set.index)
```

```python
train_set.describe()
```

```python
test_set.describe()
```

```python
label = 'GrowthRate'
train_set[label].describe()
```

```python
predictor = TabularPredictor(label=label).fit(train_set, time_limit=180)
```

```python
y_pred = predictor.predict(test_set.drop(columns=[label]))
y_pred.head()
```

```python
predictor.evaluate(test_set, silent=True)
```

```python
predictor.leaderboard(test_set)
```

```python
predictor.feature_importance(train_set)
```
