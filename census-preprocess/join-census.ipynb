```python
import pandas as pd
import os
```

```python
datasets = '../Datasets/census/data-clean'
dfs = []

for filename in os.listdir(datasets):
    if filename.endswith('.csv') and 'census-' in filename:
        year = filename.split('-')[1].split('.')[0]
        
        df = pd.read_csv(os.path.join(datasets, filename), dtype={'GEOGRAPHY': str})
        df['YEAR'] = year
        df = df[['GEOGRAPHY', 'YEAR'] + [col for col in df.columns if col not in ['GEOGRAPHY', 'YEAR']]]
        
        
        dfs.append(df)

result = pd.concat(dfs, ignore_index=True)
result = result.sort_values(by=['GEOGRAPHY', 'YEAR'])
```
