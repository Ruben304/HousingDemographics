```python
import pandas as pd
df = pd.read_csv('../Datasets/census/combined-datasets/census-percentages.csv', dtype={'GEOGRAPHY': str})
df = df[~df.apply(lambda x: x.astype(str).str.contains('-')).any(axis=1)]
```

```python
portion = df.head(20)
portion
```

```python
grouped = df.groupby('GEOGRAPHY')
resultRows = []

for name, group in grouped:
    years = list(group['YEAR'])
    yearsIndex = len(years)
    for yearIndex in range(1, len(years)):
        thisYear = years[yearIndex]
        thisDatapoint = group.iloc[yearIndex]
        prevDatapint = group.iloc[yearIndex-1]

        columns = thisDatapoint.index.tolist()
        columns = [col for col in columns if col not in ['GEOGRAPHY', 'YEAR']]

        thisRow = {
            'GEOGRAPHY': name,
            'YEAR': thisYear
        }
        
        for column in columns:
            oldValue = float(prevDatapint[column])
            newValue = float(thisDatapoint[column])

            if oldValue == newValue:
                percentageDrift = 0
            else:
                if oldValue == 0:
                    oldValue = 0.1
                percentageDrift = ((newValue - oldValue) / oldValue) * 100.0
            
            thisRow[column] = percentageDrift
        
        resultRows.append(thisRow)

resultDf = pd.DataFrame(resultRows)
resultDf.head()

            
```
